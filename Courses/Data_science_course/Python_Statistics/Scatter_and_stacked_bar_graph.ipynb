```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
xlsx_sheet = pd.read_excel('2.6.Cross-table-and-scatter-plot-exercise.xlsx')
xlsx_sheet
```

```python
background = xlsx_sheet.iloc[2,2:]
data = xlsx_sheet.iloc[3:8,2:]
task1 = xlsx_sheet.iloc[9,2:]
task2 = xlsx_sheet.iloc[10,2:]
print("background: ", background)
print("data: ", data)
print("task1: ", task1)
print("task2: ", task2)
```

```python
employed = [60,85,95,97,97,100]
unemployed = [100-employed[i] for i in range(6)]
total = [100 for i in range(6)]
print(unemployed)
age = ['18-25', '25-35', '35-45', '45-55', '55-65', '65+']
```

```python
df = pd.DataFrame(columns=['age','employed','unemployed'])
df['employed'] = employed
df['unemployed'] = unemployed
df['age'] = age
# df['total'] = total
df = df.reset_index(drop=True)
df
```

```python
g = sns.catplot(x="age", y='employed', kind='bar', palette = 'Blues_d', data=df, height=4)
```

```python
labels = df.age
employed = df.employed
unemployed = df.unemployed

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, employed, width, label='Employed')
rects2 = ax.bar(x + width/2, unemployed, width, label='Unemployed')
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Employement')
ax.set_title('Employed/Unemployed')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```
